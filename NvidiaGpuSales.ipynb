```python
import pandas as pd

df = pd.read_csv('/content/nvidia_gpu_sales_synthetic_2026_RAW.csv')

print("--- 5 BARIS PERTAMA DATA ---")
display(df.head())

print("\n--- INFORMASI STRUKTUR DATA ---")
df.info()

print("\n--- CEK DATA KOSONG (MISSING VALUES) ---")
print(df.isnull().sum())

print("\n--- STATISTIK DESKRIPTIF ---")
display(df.describe())
```

```python
df['bundle_addon'] = df['bundle_addon'].fillna('Standalone')


df['sale_date'] = pd.to_datetime(df['sale_date'])

print("--- CEK ULANG MISSING VALUES ---")
print(df.isnull().sum())

print("\n--- CEK ULANG TIPE DATA ---")
print(df.dtypes)
```

```python
def categorize_premium(pct):
    if pct <= 5.0:
        return 'Normal MSRP'
    elif pct <= 20.0:
        return 'Slight Premium'
    else:
        return 'Heavy Scalping'

df['scalping_status'] = df['price_premium_pct'].apply(categorize_premium)
print()
display(df[['gpu_model', 'price_premium_pct', 'scalping_status']].head())
```
